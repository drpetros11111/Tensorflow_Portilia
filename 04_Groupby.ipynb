# Groupby

The groupby method allows you to group rows of data together and call aggregate functions

```python
import pandas as pd
```

```python
# We will cover reading in data in a lot more detail in a later lecture!
df = pd.read_csv('/content/Universities.csv')
```

```python
# Show first N rows (N=5 by default)
df.head()
```

Now you can use the .groupby() method to group rows together based off of a **categorical** column. This column will then be reassigned to be the index.

Notice we have 2 steps:

1. Choose a categorical column to group by
2. Choose your aggregation function. Recall an aggregation function should take multiple values and return a single value (e.g. max,min, mean, std, etc...)

```python
# Step 1 simply returns a special groupby object waiting to have an aggregate method called on it!
df.groupby('Year')
```

```python
# Calculate the mean of numeric columns only and round the result
df.groupby('Year').mean(numeric_only=True).round(2)
```

```python
type(df.groupby('Year').mean(numeric_only=True))
```

```python
df.groupby('Year').mean(numeric_only = True).round(2).sort_index(ascending=False)
```

```python
df.groupby('Year').count()
```

```python
df.head()
```

```python
# Perform a groupby on multiple columns
df.groupby(['Year', 'University', 'Geography', 'Sector']).sum()
```

```python
import pandas as pd

# Assuming df is already defined as the DataFrame

# Perform a groupby on multiple columns and sum the 'Completions' column
df_grouped = df.groupby(['Year', 'University', 'Geography', 'Sector'], as_index=False).sum()

# Print the resulting DataFrame with all columns and their values
print(df_grouped)
```

```python
import pandas as pd

# Assuming df is already defined as the DataFrame

# Perform a groupby on multiple columns and sum the 'Completions' column
df_grouped = df.groupby(['Year', 'University', 'Geography', 'Sector'], as_index=False).sum().max()

# Print the resulting DataFrame with all columns and their values
print(df_grouped)
```

```python
import pandas as pd

# Assuming df is already defined as the DataFrame

# Perform a groupby on multiple columns and sum the 'Completions' column
df_grouped = df.groupby(['Year', 'University', 'Geography', 'Sector'], as_index=False).sum().std()

# Print the resulting DataFrame with all columns and their values
print(df_grouped)
```

```python
import pandas as pd

# Assuming df is already defined as the DataFrame

# Perform a groupby on multiple columns and sum the 'Completions' column
df_grouped = df.groupby(['Year', 'University', 'Geography', 'Sector'], as_index=False)['Completions'].sum() # Only include the 'Completions' column in the aggregation

# Calculate the standard deviation of the 'Completions' column
df_std = df_grouped['Completions'].std()
# Print the resulting standard deviation
print(f'the standard deviation is : {df_std}')
```

```python
import pandas as pd

# Assuming df is already defined as the DataFrame

# Perform a groupby on multiple columns and sum the 'Completions' column
df_grouped = df.groupby(['Year', 'University', 'Geography', 'Sector'], as_index=False)['Completions'].sum() # Only include the 'Completions' column in the aggregation

# Calculate the standard deviation of the 'Completions' column
df_std = df_grouped['Completions'].std()
# Print the resulting standard deviation
print(f'The standard deviation is: {round(df_std, 2)}')
# Example value, replace with your actual std value
print(f'The standard deviation is: {df_std:,.2f}')
```

## Grouping By multiple columns

```python
df.head()
```

```python
import pandas as pd

# Sample data
data = {'Sector': ['Private for-profit, 2-year', 'Private for-profit, less-than 2-year',
                   'Private for-profit, less-than 2-year', 'Private for-profit, less-than 2-year',
                   'Public, 4-year or above'],
        'University': ['Pima Medical Institute-Las Vegas', 'Healthcare Preparatory Institute',
                       'Milan Institute-Las Vegas', 'Utah College of Massage Therapy-Vegas',
                       'Western Nevada College'],
        'Year': [2016, 2016, 2016, 2016, 2016],
        'Completions': [591, 28, 408, 240, 960],
        'Geography': ['Nevada', 'Nevada', 'Nevada', 'Nevada', 'Nevada']}

df = pd.DataFrame(data)

# Grouping by 'Year' and 'Sector' and calculating the mean of 'Completions' only
df_grouped = df.groupby(['Year', 'Sector'])['Completions'].mean().round(2) #Removed round() from the groupby object and called it on the result of the mean aggregation.

# Print the result
print(df_grouped)
```

# Groupby along with other aggregate functions
This code snippet demonstrates how to create and manipulate a DataFrame in Pandas to perform a groupby operation and calculate the mean of a specific column.

----------------------

# 1. Import Pandas Library:

    import pandas as pd
Imports the Pandas library, which is essential for data manipulation and analysis.

----------------------
#2. Create Sample Data:

    data = {'Sector': ['Private for-profit, 2-year', 'Private for-profit, less-than 2-year',
                   'Private for-profit, less-than 2-year', 'Private for-profit, less-than 2-year',
                   'Public, 4-year or above'],
        'University': ['Pima Medical Institute-Las Vegas', 'Healthcare Preparatory Institute',
                       'Milan Institute-Las Vegas', 'Utah College of Massage Therapy-Vegas',
                       'Western Nevada College'],
        'Year': [2016, 2016, 2016, 2016, 2016],
        'Completions': [591, 28, 408, 240, 960],
        'Geography': ['Nevada', 'Nevada', 'Nevada', 'Nevada', 'Nevada']}

Creates a dictionary data with columns: 'Sector', 'University', 'Year', 'Completions', and 'Geography'.

Each key in the dictionary represents a column name, and the associated list contains the column values.

-----------------------------------
#3. Convert Dictionary to DataFrame:

    df = pd.DataFrame(data)

Converts the dictionary data into a Pandas DataFrame df.

The DataFrame is a tabular data structure where rows and columns are indexed, allowing for efficient data manipulation.

-------------------
#4. Group Data and Calculate Mean:

    df_grouped = df.groupby(['Year', 'Sector'])['Completions'].mean()

df.groupby(['Year', 'Sector']):

Groups the DataFrame by the 'Year' and 'Sector' columns.

Each unique combination of these columns creates a group.

['Completions']:

Selects the 'Completions' column to perform the aggregation.

.mean():

Calculates the mean of the 'Completions' column for each group.

----------------------------
#5. Print the Result:

    print(df_grouped)
Displays the resulting Series df_grouped where the index is the unique combinations of 'Year' and 'Sector', and the values are the mean completions for each group.

-------------------------
##Example Output:

Year  Sector                               
2016  Private for-profit, 2-year                 591.000000
      Private for-profit, less-than 2-year       225.333333
      Public, 4-year or above                    960.000000

Name: Completions, dtype: float64

Year and Sector are the index levels in the output.

Mean of Completions: Shows the average number of completions for each unique combination of Year and Sector.

----------------------------------
#Summary
The code creates a DataFrame from sample data, groups the data by Year and Sector, calculates the mean value of the Completions column for each group, and prints the results. This demonstrates a common data aggregation task in Pandas.
